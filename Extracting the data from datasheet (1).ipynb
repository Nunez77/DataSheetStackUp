```python
import pandas as pd
import os
```

creating the 30 data sheet of '.xlsx' format. Here each file has the value as file name in it.

```python
for i in range(30):
    file_name = 'sheet'+str(i+1)+'.xlsx'
    li = [file_name]
    li = li*6000
    df=pd.DataFrame(li)
    df.to_excel(file_name, index = False)
    
```

printing one data sheet file('sheet30.xlsx'). it has one column and 6000 rows.In every row the value is the filename here.

Here saving the 30 datasheets of folder names with filename

```python
for i in range(30):
    folder_name = os.getcwd()
    li = [folder_name+'/'+'datasheet'+str(i+1)+'.xlsx']*6000
    folder_df=pd.DataFrame(li)
    folder_df.to_excel(folder_name+'/'+'datasheet'+str(i+1)+'.xlsx', index = False)
```

printing one data sheet file('datasheet30.xlsx'). it has one column and 6000 rows.In every row the value is the folder name with file name here. 

```python
folder_df
```

Stacking all the files in column wise and saving it to .csv file.

```python
final_df = pd.DataFrame([])
for i in range(30):
    file_name = 'sheet'+str(i+1)+'.xlsx'
    temp_df =pd.read_excel(file_name)
    final_df = pd.concat([final_df,temp_df],axis=1)
final_df.to_csv('single_sheet.csv',index = False)
```

printing the final stacked dataframe below

```python
final_df
```

stacking the files which contain folder names and creating a single dataframe and saving it to '.csv' file.

```python
final_folder_df= pd.DataFrame([])
for i in range(30):
    folder_name = os.getcwd()
    file_name = folder_name+'/'+'datasheet'+str(i+1)+'.xlsx'
    temp_folder_df= pd.read_excel(file_name)
    final_folder_df = pd.concat([final_folder_df,temp_folder_df],axis=1)
final_folder_df.to_csv('single_folder_sheet.csv',index = False)
```
